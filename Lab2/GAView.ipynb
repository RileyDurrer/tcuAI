Run block below to run Genetic algorithim for 100 generations to find an opriximate maximum of the function

```python
import GAController
import matplotlib.pyplot as plt

# Generation numbers
generations = list(range(1, 101))

# Best scores of each generation
bestOf = GAController.runGenerations()

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(generations, bestOf, marker='o', linestyle='-', color='b')
plt.title('Best Score of Each Generation')
plt.xlabel('Generation')
plt.ylabel('Best of each Gen')
plt.grid(True)
plt.show()
```
